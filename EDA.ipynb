```python
df['Product'].value_counts().plot(kind='bar', figsize=(10,5), title='Product Distribution')
plt.xlabel("Product")
plt.ylabel("Number of Orders")
plt.xticks(rotation=45)
plt.savefig("product_distribution.png", dpi=100, bbox_inches="tight")
plt.show()
```

```python
df['TotalPrice'].sum()
```

```python
df['Date']=pd.to_datetime(df['Date'])
```

```python
df['Month'] = df['Date'].dt.month_name()

monthly_sales = df.groupby('Month')['TotalPrice'].sum()

monthly_sales
```

```python
monthly_sales.plot(kind='line', marker='o', figsize=(10,5))

plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales")

plt.savefig("monthly_sales.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
#outlier detection
plt.figure(figsize=(7,5))
c='b'
plt.boxplot(df['TotalPrice'],patch_artist=True,boxprops=dict(facecolor='m'))

plt.title("Outlier Detection - Total Price")

plt.savefig("outliers_totalprice.png", dpi=100, bbox_inches="tight")

plt.show()
```
